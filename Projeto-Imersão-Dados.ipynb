### 2. Obtendo os valores únicos

Método ```unique()``` retorna todos os valores únicos de uma dada coluna.

### 3. Obtendo o total de cada classe de dados

Método ```value_counts()``` retorna a quantidade total de cada valor de uma determinada coluna.

### Desafio 07: Deixar os gráficos bonitões. (Matplotlib.pyplot)

````python
import matplotlib.pyplot as plt

plt.bar(df['tratamento'].unique(),df['tratamento'].value_counts(), color ='#ffa381')
plt.title('Tipos de tratamento', size=16)
plt.xlabel("tratamento", size=12)
plt.ylabel("count", size=12)
plt.show()
````

````python
import seaborn as sns
plt.figure(figsize=(10,5))
plt.title('Tempos aplicados', size=20)
sns.countplot(data=df,x='tempo', palette ='pastel');
````
